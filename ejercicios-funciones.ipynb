## Ejercicio P1 — Perímetro de un rectángulo
Define una función calcularPerimetro que reciba la base y la altura de un rectángulo y devuelva su perímetro.

Pruébala con los siguientes valores:

base=6, altura=4   → 20
base=10, altura=3  → 26

````scala
def calcularPerimetro(base: Int, altura: Int): Int = (base + altura) * 2

val perimetro1 = calcularPerimetro(6, 4)
val perimetro2 = calcularPerimetro(10, 3)

println(s"base=6, altura=4   → $perimetro1")
println(s"base=10, altura=3  → $perimetro2")
````

## Ejercicio P2 — Mayor de edad
Define una función esMayorDeEdad que reciba una edad (Int) y devuelva un Boolean indicando si la persona es mayor de edad (18 o más).

Pruébala con:

esMayorDeEdad(17)  → false
esMayorDeEdad(18)  → true
esMayorDeEdad(25)  → true

````scala
def esMayorDeEdad(edad: Int): Boolean = edad >= 18

println(s"esMayorDeEdad(17)  → ${esMayorDeEdad(17)}")
println(s"esMayorDeEdad(18)  → ${esMayorDeEdad(18)}")
println(s"esMayorDeEdad(25)  → ${esMayorDeEdad(25)}")
````

## Ejercicio P3 — Conversión de euros a dólares
Define una función euroADolar que convierta una cantidad en euros a dólares usando el tipo de cambio 1.08.

Pruébala con:

euroADolar(100.0)  → 108.0
euroADolar(250.0)  → 270.0

````scala
def euroADolar(euros: Double): Double = euros * 1.08

println(s"euroADolar(100.0)  → ${euroADolar(100.0)}")
println(s"euroADolar(250.0)  → ${euroADolar(250.0)}")
````

## Ejercicio P4 — Saludo multiidioma
Define una función saludo con dos parámetros:

nombre: String — obligatorio
idioma: String — con valor por defecto "es"
Si idioma es "es" devuelve "Hola, <nombre>!", si es "en" devuelve "Hello, <nombre>!".

Pruébala con:

saludo("Ana")           → "Hola, Ana!"
saludo("Ana", "en")     → "Hello, Ana!"

````scala
def saludo(nombre: String, idioma: String = "es"): String =
  if (idioma == "es") s"Hola, $nombre!"
  else if (idioma == "en") s"Hello, $nombre!"
  else s"Hola, $nombre!"

println(saludo("Ana"))
println(saludo("Ana", "en"))
````

## Ejercicio P5 — Conversión de temperaturas
Dada la siguiente lista:

```scala
val temperaturas = List(18.0, 22.5, 30.1, 15.3, 27.8)
```

Usa map para convertir todas las temperaturas de Celsius a Fahrenheit con la fórmula `(c * 9.0 / 5.0) + 32`.

Salida esperada:

```
List(64.4, 72.5, 86.18, 59.54, 82.04)
```

````scala
val temperaturas = List(18.0, 22.5, 30.1, 15.3, 27.8)

val fahrenheit = temperaturas.map(c => (c * 9.0 / 5.0) + 32)

println(fahrenheit)
````

## Ejercicio P6 — Filtrar palabras por longitud

Usa filter para quedarte solo con las palabras que tengan más de 5 letras.

List(hadoop, kafka)

## Ejercicio P7 — Encontrar el mínimo con reduce
Dada la siguiente lista:

```scala
val numeros = List(3, 7, 2, 9, 4, 6)
```

Usa reduce para encontrar el número mínimo de la lista.

Salida esperada:

```
2
```

````scala
val numeros = List(3, 7, 2, 9, 4, 6)

val minimo = numeros.reduce((a, b) => if (a < b) a else b)

println(minimo)
````

## Ejercicio P8 — Función al cuadrado (tres estilos)
Implementa la misma función de tres formas distintas.

La función recibe un número `Int` y devuelve ese número multiplicado por sí mismo (al cuadrado).

```scala
// 1. Con def
def cuadrado(x: Int): Int = x * x

// 2. Anónima pura — se ejecuta directamente con el valor 7
((x: Int) => x * x)(7)

// 3. Guardada en val
val cuadrado = (x: Int) => x * x
```

Salida esperada en los tres casos para el valor 7:

```
49
```

````scala
// 1. Con def
def cuadrado(x: Int): Int = x * x
println(s"def cuadrado(7) = ${cuadrado(7)}")

// 2. Anónima pura — se ejecuta directamente con el valor 7
val resultAnon = ((x: Int) => x * x)(7)
println(s"Anónima pura: $resultAnon")

// 3. Guardada en val
val cuadradoFunc = (x: Int) => x * x
println(s"val cuadradoFunc(7) = ${cuadradoFunc(7)}")
````

## Ejercicio P9 — Concatenar strings (tres estilos)
Implementa la misma función de tres formas distintas.

La función recibe dos `String` y los concatena separados por un espacio.

```scala
// 1. Con def
def concatenar(a: String, b: String): String = a + " " + b

// 2. Anónima pura — pruébala con ("Big", "Data")
((a: String, b: String) => a + " " + b)("Big", "Data")

// 3. Guardada en val
val concatenar = (a: String, b: String) => a + " " + b
```

Salida esperada para ("Big", "Data"):

```
"Big Data"
```

````scala
// 1. Con def
def concatenar(a: String, b: String): String = a + " " + b
println(s"def concatenar: ${concatenar("Big", "Data")}")

// 2. Anónima pura — pruébala con ("Big", "Data")
val resultConcat = ((a: String, b: String) => a + " " + b)("Big", "Data")
println(s"Anónima pura: $resultConcat")

// 3. Guardada en val
val concatenarFunc = (a: String, b: String) => a + " " + b
println(s"val concatenarFunc: ${concatenarFunc("Big", "Data")}")
````

## Ejercicio P10 — Filtrar productos por precio
Dada la siguiente lista de tuplas (producto, precio):

```scala
val productos = List(
  ("Teclado",  45.0),
  ("Ratón",    20.0),
  ("Monitor", 180.0),
  ("Webcam",   35.0),
  ("Auriculares", 60.0)
)
```

Extrae solo los nombres de los productos con un precio superior a 40.0.

Salida esperada:

```
List(Teclado, Monitor, Auriculares)
```

````scala
val productos = List(
  ("Teclado",  45.0),
  ("Ratón",    20.0),
  ("Monitor", 180.0),
  ("Webcam",   35.0),
  ("Auriculares", 60.0)
)

val productosCatos = productos
  .filter { case (nombre, precio) => precio > 40.0 }
  .map { case (nombre, precio) => nombre }

println(productosCatos)
````
