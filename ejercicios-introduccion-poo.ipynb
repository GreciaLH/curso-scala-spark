### Ejercicio P2 — Clase Rectangulo

Crea una clase `Rectangulo` con los atributos `base` (Double) y `altura` (Double).

Añade los métodos:
- `area()` → devuelve base * altura
- `perimetro()` → devuelve 2 * (base + altura)
- `esCuadrado()` → devuelve true si base y altura son iguales

Crea tres rectángulos distintos y muestra por pantalla sus medidas, área, perímetro y si son cuadrados.

````scala
class Rectangulo(val base: Double, val altura: Double) {
  def area(): Double = {
    base * altura
  }
  
  def perimetro(): Double = {
    2 * (base + altura)
  }
  
  def esCuadrado(): Boolean = {
    base == altura
  }
  
  def mostrar(): String = {
    s"Rectángulo(base=$base, altura=$altura) - Área: ${area()}, Perímetro: ${perimetro()}, ¿Es cuadrado? ${esCuadrado()}"
  }
}

// Crear tres rectángulos distintos
val rect1 = new Rectangulo(5.0, 10.0)
val rect2 = new Rectangulo(4.0, 4.0)
val rect3 = new Rectangulo(3.5, 7.0)

// Mostrar sus medidas, área, perímetro y si son cuadrados
println(rect1.mostrar())
println(rect2.mostrar())
println(rect3.mostrar())
````

### Ejercicio P3 — Clase Bombilla

Crea una clase `Bombilla` con un atributo mutable `encendida` (Boolean) que empieza en false. 

Añade los métodos:
- `encender()` → cambia `encendida` a true e imprime "💡 Bombilla encendida."
- `apagar()` → cambia `encendida` a false e imprime "🌑 Bombilla apagada."
- `estado()` → devuelve "encendida" o "apagada" según corresponda

Crea una bombilla, enciéndela, consulta su estado, apágala y vuelve a consultar.

````scala
class Bombilla {
  var encendida: Boolean = false
  
  def encender(): Unit = {
    encendida = true
    println("💡 Bombilla encendida.")
  }
  
  def apagar(): Unit = {
    encendida = false
    println("🌑 Bombilla apagada.")
  }
  
  def estado(): String = {
    if (encendida) "encendida" else "apagada"
  }
}

// Crear una bombilla
val bombilla = new Bombilla()

// Encenderla
bombilla.encender()

// Consultar su estado
println(s"Estado actual: ${bombilla.estado()}")

// Apagarla
bombilla.apagar()

// Consultar su estado nuevamente
println(s"Estado actual: ${bombilla.estado()}")
````

### Ejercicio P4 — Clase Estudiante

Crea una clase `Estudiante` con los atributos `nombre` (String), `curso` (String) y `nota` (Double).

Añade los métodos:
- `aprobado()` → devuelve true si la nota es mayor o igual a 5.0
- `calificacion()` → devuelve "Aprobado" o "Suspenso" según la nota
- `ficha()` → devuelve una línea con nombre, curso, nota y calificación

````scala
class Estudiante(val nombre: String, val curso: String, val nota: Double) {
  def aprobado(): Boolean = {
    nota >= 5.0
  }
  
  def calificacion(): String = {
    if (aprobado()) "Aprobado" else "Suspenso"
  }
  
  def ficha(): String = {
    s"$nombre - Curso: $curso - Nota: $nota - ${calificacion()}"
  }
}

// Crear una lista con 4 estudiantes de distintas notas
val estudiantes = List(
  new Estudiante("Ana García", "2º DAM", 8.5),
  new Estudiante("Carlos López", "2º DAM", 4.2),
  new Estudiante("María Rodríguez", "2º DAM", 7.0),
  new Estudiante("Juan Martínez", "2º DAM", 3.8)
)

// Imprimir la ficha de cada uno
println("=== FICHAS DE ESTUDIANTES ===")
estudiantes.foreach(est => println(est.ficha()))
````

### Ejercicio P5 — object Calculadora

Crea un `object` llamado `Calculadora` con los métodos:
- `sumar(a: Double, b: Double)`
- `restar(a: Double, b: Double)`
- `multiplicar(a: Double, b: Double)`
- `dividir(a: Double, b: Double)` → si b == 0 imprime "Error: división por cero" y devuelve 0.0

Prueba cada operación con al menos dos llamadas, incluyendo una división por cero.

````scala
object Calculadora {
  def sumar(a: Double, b: Double): Double = {
    a + b
  }
  
  def restar(a: Double, b: Double): Double = {
    a - b
  }
  
  def multiplicar(a: Double, b: Double): Double = {
    a * b
  }
  
  def dividir(a: Double, b: Double): Double = {
    if (b == 0) {
      println("Error: división por cero")
      0.0
    } else {
      a / b
    }
  }
}

// Pruebas de cada operación
println("=== PRUEBAS DE CALCULADORA ===")
println(s"Suma: 10 + 5 = ${Calculadora.sumar(10, 5)}")
println(s"Suma: 3.5 + 2.5 = ${Calculadora.sumar(3.5, 2.5)}")

println(s"Resta: 10 - 5 = ${Calculadora.restar(10, 5)}")
println(s"Resta: 8.5 - 3.2 = ${Calculadora.restar(8.5, 3.2)}")

println(s"Multiplicación: 10 * 5 = ${Calculadora.multiplicar(10, 5)}")
println(s"Multiplicación: 4.5 * 2 = ${Calculadora.multiplicar(4.5, 2)}")

println(s"División: 10 / 5 = ${Calculadora.dividir(10, 5)}")
println(s"División: 15 / 3 = ${Calculadora.dividir(15, 3)}")
println("División por cero: 10 / 0 = " + Calculadora.dividir(10, 0))
````

### Ejercicio P7 — Clase Temporizador

Crea una clase `Temporizador` con un atributo mutable `segundos` (Int) que recibe el valor inicial en el constructor.

Añade los métodos:
- `avanzar(n: Int)` → suma n segundos al contador
- `reiniciar()` → pone segundos a 0
- `mostrarTiempo()` → imprime el tiempo en formato "mm:ss" (ej: "02:05")

Pista: minutos = segundos / 60, segundos restantes = segundos % 60. Usa `f"${min}%02d:${seg}%02d"` para el formato.

Crea un temporizador, avanza varios intervalos, muestra el tiempo y reinícialo.

````scala
class Temporizador(var segundos: Int) {
  def avanzar(n: Int): Unit = {
    segundos += n
  }
  
  def reiniciar(): Unit = {
    segundos = 0
  }
  
  def mostrarTiempo(): Unit = {
    val min = segundos / 60
    val seg = segundos % 60
    println(f"$min%02d:$seg%02d")
  }
}

// Crear un temporizador con valor inicial
println("=== PRUEBAS DE TEMPORIZADOR ===")
val temp = new Temporizador(0)

// Avanzar varios intervalos
println("Tiempo inicial:")
temp.mostrarTiempo()

println("\nAvanzar 65 segundos:")
temp.avanzar(65)
temp.mostrarTiempo()

println("\nAvanzar 125 segundos más:")
temp.avanzar(125)
temp.mostrarTiempo()

println("\nAvanzar 30 segundos más:")
temp.avanzar(30)
temp.mostrarTiempo()

println("\nReiniciar temporizador:")
temp.reiniciar()
temp.mostrarTiempo()
````

### Ejercicio P8 — Clase CajaRegistradora

Crea una clase `CajaRegistradora` con los atributos `tienda` (String) y un atributo privado mutable `totalVentas` (Double) inicializado a 0.0.

Añade los métodos:
- `registrarVenta(importe: Double)` → suma el importe a totalVentas e imprime el ticket
- `obtenerTotal()` → devuelve totalVentas
- `cerrarCaja()` → imprime el total del día y pone totalVentas a 0.0

Simula 4 ventas, consulta el total y cierra la caja.

````scala
class CajaRegistradora(val tienda: String) {
  private var totalVentas: Double = 0.0
  
  def registrarVenta(importe: Double): Unit = {
    totalVentas += importe
    val impStr = "%.2f".format(importe)
    val totalStr = "%.2f".format(totalVentas)
    println(s"Tienda: $tienda - Ticket: $$$impStr - Total: $$$totalStr")
  }
  
  def obtenerTotal(): Double = {
    totalVentas
  }
  
  def cerrarCaja(): Unit = {
    println(s"=== CIERRE DE CAJA ===")
    println(s"Tienda: $tienda")
    val totalStr = "%.2f".format(totalVentas)
    println(s"Total del día: $$$totalStr")
    totalVentas = 0.0
    println("Caja reiniciada.")
  }
}

// Simular 4 ventas
println("=== OPERACIONES DE CAJA REGISTRADORA ===\n")
val caja = new CajaRegistradora("SuperMercado Central")

caja.registrarVenta(25.50)
caja.registrarVenta(12.30)
caja.registrarVenta(45.75)
caja.registrarVenta(18.99)

val totalStr = "%.2f".format(caja.obtenerTotal())
println(s"\nTotal acumulado: $$$totalStr")

// Cerrar la caja
println()
caja.cerrarCaja()
val totalStr2 = "%.2f".format(caja.obtenerTotal())
println(s"Total después de cerrar: $$$totalStr2")
````

### Ejercicio P9 — Clase Cancion y object Reproductor

Crea una clase `Cancion` con los atributos `titulo` (String), `artista` (String) y `duracionSegundos` (Int). Añade un método `duracionFormateada()` que devuelva la duración en formato "mm:ss".

Crea un `object` llamado `Reproductor` con:
- Un atributo mutable `volumen` (Int) inicializado a 50
- Un método `reproducir(c: Cancion)` que imprima qué canción suena y a qué volumen
- Un método `subirVolumen(n: Int)` que sume sin superar 100
- Un método `bajarVolumen(n: Int)` que reste sin bajar de 0

Crea tres canciones, reprodúcelas y cambia el volumen entre reproducciones.

````scala
class Cancion(val titulo: String, val artista: String, val duracionSegundos: Int) {
  def duracionFormateada(): String = {
    val min = duracionSegundos / 60
    val seg = duracionSegundos % 60
    f"$min%02d:$seg%02d"
  }
}

object Reproductor {
  var volumen: Int = 50
  
  def reproducir(c: Cancion): Unit = {
    println(s"🎵 Reproduciendo: ${c.titulo} de ${c.artista}")
    println(s"   Duración: ${c.duracionFormateada()} - Volumen: $volumen%")
  }
  
  def subirVolumen(n: Int): Unit = {
    volumen = scala.math.min(volumen + n, 100)
    println(s"Volumen aumentado a: $volumen%")
  }
  
  def bajarVolumen(n: Int): Unit = {
    volumen = scala.math.max(volumen - n, 0)
    println(s"Volumen reducido a: $volumen%")
  }
}

// Crear tres canciones
val cancion1 = new Cancion("Bohemian Rhapsody", "Queen", 354)
val cancion2 = new Cancion("Imagine", "John Lennon", 183)
val cancion3 = new Cancion("Hotel California", "Eagles", 391)

// Reproducir canciones y cambiar volumen
println("=== REPRODUCTOR DE MÚSICA ===\n")

Reproductor.reproducir(cancion1)

println("\nSubiendo volumen...")
Reproductor.subirVolumen(25)

println("\nReproduciendo siguiente canción:")
Reproductor.reproducir(cancion2)

println("\nBajando volumen...")
Reproductor.bajarVolumen(30)

println("\nReproduciendo tercera canción:")
Reproductor.reproducir(cancion3)

println("\nAumentando volumen al máximo...")
Reproductor.subirVolumen(50)
````

### Ejercicio P10 — Clase Factura con constructor secundario y encapsulamiento

Crea una clase `Factura` con los atributos `numeroFactura` (Int), `cliente` (String), `baseImponible` (Double) y `tipoIVA` (Double).

Añade un constructor secundario que solo reciba `numeroFactura`, `cliente` y `baseImponible`, usando un IVA del 21% por defecto.

Añade los métodos (todos con lógica private donde corresponda):
- `private def calcularIVA()` → devuelve baseImponible * tipoIVA / 100
- `total()` → devuelve baseImponible + calcularIVA()
- `imprimir()` → muestra un resumen con número, cliente, base, IVA y total

Crea dos facturas: una con IVA personalizado (10%) y otra usando el constructor secundario.

````scala
class Factura(val numeroFactura: Int, val cliente: String, val baseImponible: Double, val tipoIVA: Double) {
  // Constructor secundario con IVA del 21% por defecto
  def this(numeroFactura: Int, cliente: String, baseImponible: Double) = {
    this(numeroFactura, cliente, baseImponible, 21.0)
  }
  
  // Método privado para calcular IVA
  private def calcularIVA(): Double = {
    baseImponible * tipoIVA / 100
  }
  
  // Método público que devuelve el total
  def total(): Double = {
    baseImponible + calcularIVA()
  }
  
  // Método para imprimir el resumen
  def imprimir(): Unit = {
    val iva = calcularIVA()
    val totalVal = total()
    val baseStr = "%.2f".format(baseImponible)
    val ivaStr = "%.2f".format(iva)
    val totalStr = "%.2f".format(totalVal)
    
    println(s"=== FACTURA #$numeroFactura ===")
    println(s"Cliente: $cliente")
    println(s"Base Imponible: $$$baseStr")
    println(s"IVA ($tipoIVA%): $$$ivaStr")
    println(s"Total: $$$totalStr")
    println()
  }
}

// Crear dos facturas
println("=== GESTIÓN DE FACTURAS ===\n")

// Factura 1: con IVA personalizado (10%)
val factura1 = new Factura(1001, "Empresa A", 1000.0, 10.0)
factura1.imprimir()

// Factura 2: usando constructor secundario (IVA 21% por defecto)
val factura2 = new Factura(1002, "Empresa B", 500.0)
factura2.imprimir()

// Mostrar totales
val total1Str = "%.2f".format(factura1.total())
val total2Str = "%.2f".format(factura2.total())
println(s"Total Factura #1001: $$$total1Str")
println(s"Total Factura #1002: $$$total2Str")
````
