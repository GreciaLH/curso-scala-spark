## Ejercicio P1 — Géneros únicos de una biblioteca

Una biblioteca registra los géneros literarios disponibles en su catálogo. El problema es que la base de datos tiene entradas repetidas porque distintos bibliotecarios los fueron añadiendo sin coordinación.

Dada esta lista:

```scala
val generosRepetidos = List("Novela", "Poesía", "Ensayo", "Novela", "Teatro", "Poesía", "Cómic", "Novela", "Ensayo", "Cómic")
```

- Convierte la lista en un `Set` para obtener los géneros únicos.
- Imprime cuántos géneros únicos hay (usa `.size` o `.length`).
- Muestra el listado ordenado alfabéticamente.

**Salida esperada:**

```
Géneros únicos: 5
Listado: List(Cómic, Ensayo, Novela, Poesía, Teatro)
```

````scala
val generosRepetidos = List("Novela", "Poesía", "Ensayo", "Novela", "Teatro", "Poesía", "Cómic", "Novela", "Ensayo", "Cómic")

// Convierte la lista en un Set para obtener los géneros únicos
val generosUnicos = generosRepetidos.toSet

// Imprime cuántos géneros únicos hay
println(s"Géneros únicos: ${generosUnicos.size}")

// Muestra el listado ordenado alfabéticamente
val generosOrdenados = generosUnicos.toList.sorted
println(s"Listado: $generosOrdenados")
````

## Ejercicio P2 — Catálogo rápido con Map

Una tienda de informática necesita un catálogo rápido de consulta. Crea un `Map[String, Double]` con los siguientes productos y precios:

```scala
"Teclado" -> 45.99
"Ratón" -> 22.50
"Webcam" -> 67.00
"Auriculares" -> 89.95
```

Realiza estas tres consultas e imprime el resultado de cada una:

- El precio del `"Ratón"`.
- El precio de `"Monitor"` usando `.getOrElse` con valor por defecto `0.0` (porque el producto no existe).

**Salida esperada:**

```
Precio Ratón: 22.5
```

## Ejercicio P3 — Función que devuelve extremos en una tupla

Escribe una función llamada `extremos` que:

- Reciba un parámetro `List[Int]`
- Devuelva una `Tuple2[Int, Int]` con dos elementos:
  - Primer elemento: el valor **mínimo** de la lista
  - Segundo elemento: el valor **máximo** de la lista

Prueba la función con esta lista:

```scala
List(34, 7, 89, 12, 56, 3, 71)
```

Muestra el resultado desestructurando la tupla en dos variables llamadas `menor` y `mayor`, e imprime ambos valores.

**Salida esperada:**

```
Menor: 3
Mayor: 89
```

````scala
// Define la función extremos que devuelve una Tuple2[Int, Int]
def extremos(lista: List[Int]): Tuple2[Int, Int] = {
  (lista.min, lista.max)
}

// Prueba la función con la lista proporcionada
val numeros = List(34, 7, 89, 12, 56, 3, 71)
val resultado = extremos(numeros)

// Desestructura la tupla en dos variables
val (menor, mayor) = resultado

// Imprime los resultados
println(s"Menor: $menor")
println(s"Mayor: $mayor")
````

## Ejercicio P4 — Sistema de votaciones con Map

Un sistema de votaciones recoge los votos emitidos en una asamblea. Tienes el siguiente `Map` con los resultados ya contados:

```scala
var votos = Map("Ana" -> 187, "Carlos" -> 198, "Luis" -> 74, "Marta" -> 231)
```

Realiza estas **cuatro operaciones en orden**:

1. Añade un nuevo candidato `"Elena"` con `102` votos (usa `+` o `+=()`).
2. Actualiza los votos de `"Carlos"` a `215` (usa `updated()`).
3. Elimina al candidato `"Luis"` (usa `-` o `--=()`).
4. Muestra el mapa final con los candidatos **ordenados alfabéticamente**.

**Salida esperada:**

```
Resultados finales:
Ana → 187
Carlos → 215
Elena → 102
Marta → 231
```

````scala
var votos = Map("Ana" -> 187, "Carlos" -> 198, "Luis" -> 74, "Marta" -> 231)

// 1. Añade un nuevo candidato "Elena" con 102 votos
votos = votos + ("Elena" -> 102)

// 2. Actualiza los votos de "Carlos" a 215
votos = votos.updated("Carlos", 215)

// 3. Elimina al candidato "Luis"
votos = votos - "Luis"

// 4. Muestra el mapa final con los candidatos ordenados alfabéticamente
println("Resultados finales:")
votos.toList.sortBy(_._1).foreach { case (candidato, cantidad) =>
  println(s"$candidato → $cantidad")
}
````

## Ejercicio P5 — Clasificación de estudiantes por edad con tuplas

Crea una `List` de `Tuple2[String, Int]` que represente el nombre y la edad de cinco estudiantes:

```scala
val estudiantes = List(
  ("Laura", 19),
  ("Marcos", 25),
  ("Irene", 31),
  ("Pablo", 22),
  ("Rosa", 17)
)
```

Recorre la lista con un **`for` loop** y desstructura cada tupla en `(nombre, edad)`. Para cada estudiante, imprime un mensaje distinto según su edad:

- Si tiene **menos de 20 años**: `"Estudiante joven"`
- Si tiene **entre 20 y 30 años** (inclusive): `"Estudiante adulto"`
- Si tiene **más de 30 años**: `"Estudiante senior"`

**Salida esperada:**

```
Laura (19) → Estudiante joven
Marcos (25) → Estudiante adulto
Irene (31) → Estudiante senior
Pablo (22) → Estudiante adulto
Rosa (17) → Estudiante joven
```

````scala
val estudiantes = List(
  ("Laura", 19),
  ("Marcos", 25),
  ("Irene", 31),
  ("Pablo", 22),
  ("Rosa", 17)
)

// Recorre la lista con un for loop y desstructura las tuplas
for ((nombre, edad) <- estudiantes) {
  val categoria = if (edad < 20) {
    "Estudiante joven"
  } else if (edad >= 20 && edad <= 30) {
    "Estudiante adulto"
  } else {
    "Estudiante senior"
  }
  
  println(s"$nombre ($edad) → $categoria")
}
````

## Ejercicio P6 — Operaciones de conjuntos en clínica veterinaria

Una clínica veterinaria lleva el control de las vacunas administradas a sus pacientes. Tienes dos `Set[String]` con los nombres de los animales que han recibido cada vacuna:

```scala
val vacunaA = Set("Toby", "Luna", "Max", "Nala", "Rocky")
val vacunaB = Set("Luna", "Rocky", "Bella", "Max", "Coco")
```

Calcula e imprime:

- Los animales que han recibido **ambas vacunas** (intersección: `&` o `.intersect`)
- Los animales que **solo han recibido la vacuna A** (diferencia: `--`)
- Los animales que **solo han recibido la vacuna B** (diferencia: `--`)
- El **total de animales únicos** atendidos en la clínica (unión: `|` o `.union`)

**Salida esperada:**

```
Recibieron ambas vacunas: Set(Luna, Max, Rocky)
Solo vacuna A: Set(Toby, Nala)
Solo vacuna B: Set(Bella, Coco)
Total animales atendidos: 7
```

````scala
val vacunaA = Set("Toby", "Luna", "Max", "Nala", "Rocky")
val vacunaB = Set("Luna", "Rocky", "Bella", "Max", "Coco")

// Animales que han recibido ambas vacunas (intersección)
val ambas = vacunaA & vacunaB
println(s"Recibieron ambas vacunas: $ambas")

// Animales que solo han recibido la vacuna A
val soloA = vacunaA -- vacunaB
println(s"Solo vacuna A: $soloA")

// Animales que solo han recibido la vacuna B
val soloB = vacunaB -- vacunaA
println(s"Solo vacuna B: $soloB")

// Total de animales únicos atendidos en la clínica (unión)
val totalAnimales = vacunaA | vacunaB
println(s"Total animales atendidos: ${totalAnimales.size}")
````

## Ejercicio P7 — Consulta de envíos con Map

Una empresa de mensajería registra el estado de sus envíos en un `Map[String, String]` donde:
- **Clave**: código del envío (ej: `"ENV-001"`)
- **Valor**: estado actual (ej: `"En tránsito"`)

```scala
val envios = Map(
  "ENV-001" -> "En tránsito",
  "ENV-002" -> "Entregado",
  "ENV-003" -> "En almacén",
  "ENV-004" -> "Devuelto"
)
```

Escribe una función llamada `consultarEnvio` que:

- Reciba el mapa y un código de envío como parámetros
- Devuelva el **estado si existe** el código
- Devuelva el mensaje **`"Código no encontrado"`** si no existe el código

Prueba la función con **al menos un código válido** y **uno inválido**.

**Salida esperada:**

```
ENV-002: Entregado
ENV-007: Código no encontrado
```

````scala
val envios = Map(
  "ENV-001" -> "En tránsito",
  "ENV-002" -> "Entregado",
  "ENV-003" -> "En almacén",
  "ENV-004" -> "Devuelto"
)

// Función para consultar el estado de un envío
def consultarEnvio(mapa: Map[String, String], codigo: String): String = {
  mapa.getOrElse(codigo, "Código no encontrado")
}

// Prueba con un código válido
val resultado1 = consultarEnvio(envios, "ENV-002")
println(s"ENV-002: $resultado1")

// Prueba con un código inválido
val resultado2 = consultarEnvio(envios, "ENV-007")
println(s"ENV-007: $resultado2")
````

## Ejercicio P8 — Contador de frecuencia de palabras con groupBy

Tienes una lista de palabras extraídas de un documento. Necesitas contar cuántas veces aparece cada palabra.

```scala
val palabras = List(
  "Scala", "es", "un", "lenguaje", "scala", "funcional", "un", 
  "lenguaje", "muy", "expresivo", "es", "muy", "potente"
)
```

Realiza las siguientes operaciones:

1. **Normaliza** todas las palabras a minúsculas con `.toLowerCase` (para que `"Scala"` y `"scala"` cuenten como la misma palabra)
2. **Agrupa** las palabras usando `.groupBy(palabra => palabra)`
3. **Cuenta** la frecuencia con `.mapValues(_.length)` (o `.mapValues(_.size)`)
4. **Filtra** para quedarte solo con las palabras que aparecen **más de una vez**
5. **Ordena** de mayor a menor frecuencia usando `.sortBy(-_._2)`
6. **Imprime** las palabras repetidas en formato: `palabra → frecuencia`

**Salida esperada:**

```
Palabras repetidas:
es → 2
lenguaje → 2
muy → 2
scala → 2
un → 2
```

````scala
val palabras = List(
  "Scala", "es", "un", "lenguaje", "scala", "funcional", "un", 
  "lenguaje", "muy", "expresivo", "es", "muy", "potente"
)

// Convierte a minúsculas, agrupa y cuenta frecuencias
val frecuencias = palabras
  .map(_.toLowerCase)
  .groupBy(palabra => palabra)
  .view.mapValues(_.length).toMap

// Filtra palabras que aparecen más de una vez y ordena por frecuencia descendente
val repetidas = frecuencias
  .filter(_._2 > 1)
  .toList
  .sortBy(-_._2)

// Imprime las palabras repetidas
println("Palabras repetidas:")
repetidas.foreach { case (palabra, frecuencia) =>
  println(s"$palabra → $frecuencia")
}
````

## Ejercicio P9 — Sistema de matrículas con Map[String, Set[String]]

Un instituto registra las asignaturas en las que está matriculado cada alumno mediante un `Map[String, Set[String]]`:

```scala
val matriculas = Map(
  "Ana" -> Set("Matemáticas", "Física", "Inglés", "Historia"),
  "Carlos" -> Set("Matemáticas", "Química", "Inglés", "Educación Física"),
  "Marta" -> Set("Filosofía", "Historia", "Inglés", "Arte"),
  "Luis" -> Set("Matemáticas", "Física", "Química", "Tecnología")
)
```

Realiza las siguientes consultas:

1. **Asignaturas comunes**: Imprime las asignaturas que cursan tanto "Ana" como "Carlos" (intersección de sus conjuntos con `&`)
2. **Total de asignaturas únicas**: Calcula el total de asignaturas únicas impartidas en el instituto (unión de todos los conjuntos)
3. **Alumnos por asignatura**: Encuentra qué alumnos cursan "Matemáticas"

**Salida esperada:**

```
Asignaturas comunes Ana y Carlos: Set(Matemáticas, Inglés)
Total asignaturas únicas en el instituto: 10
Alumnos que cursan Matemáticas: List(Ana, Carlos, Luis)
```

````scala
val matriculas = Map(
  "Ana" -> Set("Matemáticas", "Física", "Inglés", "Historia"),
  "Carlos" -> Set("Matemáticas", "Química", "Inglés", "Educación Física"),
  "Marta" -> Set("Filosofía", "Historia", "Inglés", "Arte"),
  "Luis" -> Set("Matemáticas", "Física", "Química", "Tecnología")
)

// 1. Asignaturas comunes entre Ana y Carlos (intersección)
val commonAnaCarlos = matriculas("Ana") & matriculas("Carlos")
println(s"Asignaturas comunes Ana y Carlos: $commonAnaCarlos")

// 2. Total de asignaturas únicas en el instituto (unión de todos los sets)
val todasAsignaturas = matriculas.values.foldLeft(Set[String]())(_ | _)
println(s"Total asignaturas únicas en el instituto: ${todasAsignaturas.size}")

// 3. Alumnos que cursan Matemáticas
val alumnosMates = matriculas
  .filter { case (alumno, asignaturas) => asignaturas.contains("Matemáticas") }
  .keys
  .toList
  .sorted
println(s"Alumnos que cursan Matemáticas: $alumnosMates")
````

## Ejercicio P10 — Clasificación de números en tupla triple

Escribe una función llamada `clasificarNumeros` que:

- Reciba un parámetro `List[Int]`
- Devuelva una `Tuple3[List[Int], List[Int], List[Int]]` con tres listas en este orden:
  - Primer elemento: los números **negativos**
  - Segundo elemento: los números **ceros**
  - Tercer elemento: los números **positivos**

Prueba la función con esta lista:

```scala
val numeros = List(-5, 0, 3, -1, 7, 0, -9, 4, 0, 2, -3, 8)
```

Desestructura la tupla en tres variables llamadas `negativos`, `ceros` y `positivos`, e imprime cada grupo.

**Salida esperada:**

```
Negativos: List(-5, -1, -9, -3)
Ceros: List(0, 0, 0)
Positivos: List(3, 7, 4, 2, 8)
```

````scala
// Define la función clasificarNumeros
def clasificarNumeros(lista: List[Int]): Tuple3[List[Int], List[Int], List[Int]] = {
  val negativos = lista.filter(_ < 0)
  val ceros = lista.filter(_ == 0)
  val positivos = lista.filter(_ > 0)
  (negativos, ceros, positivos)
}

// Prueba la función con la lista proporcionada
val numeros = List(-5, 0, 3, -1, 7, 0, -9, 4, 0, 2, -3, 8)
val resultado = clasificarNumeros(numeros)

// Desestructura la tupla en tres variables
val (negativos, ceros, positivos) = resultado

// Imprime cada grupo
println(s"Negativos: $negativos")
println(s"Ceros: $ceros")
println(s"Positivos: $positivos")
````
